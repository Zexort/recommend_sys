```python
import json
with open('file_paintings.json', 'r') as f:
    paintings = json.load(f)['paintings']
```

```python
while True:
    pic_num = input('\nВведите номер картины: ')
    t = int(input('Введите время: '))
    pic_type = ''
    recommendations = []
    numbers = []
    dist = []

    if t >= 15:
        for i in paintings: # Поиск номера, типа, координат и автора картины
            if i['number'] == pic_num:
                pic_type = i['type']
                x = i['coord']
                aut = i['author']


        for i in paintings: # Выявление картин для рекомендации
            y = i['coord']
            abs(x[0] - y[0]) <= 2
            if (i['type'] == pic_type and i['number'] != pic_num and abs(x[0] - y[0]) <= 2) or \
               (aut == i['author'] and i['number'] != pic_num and abs(x[0] - y[0]) <= 2):
                recommendations.append(i['name'])
                numbers.append(i['number'])
                dist.append(abs(x[0] - y[0]))
                


        if len(recommendations) > 1: 
            print('Рекомендуем посмотреть эти картины:')
        else: print('Рекомендуем посмотреть эту картину:')


        for j in range(len(recommendations)): # вывод названия и номера картины
            print(recommendations[j] + ', Номер картины:', numbers[j] + ', расстояние до картины:', dist[j])
        recommendations.clear()
        numbers.clear()
        ex = input('\nПродолжаем?(Да\Нет): ')
        if ex == 'Нет': break
```
